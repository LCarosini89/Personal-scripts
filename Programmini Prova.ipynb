# Programmino che calcola la focale di una lente 


Il programmino chiede in input all'utente i raggi di curvatura della lente e restituisce la focale secondo la formula\
<br>
<br>
$$\frac{1}{f} = (n-1) \left( \frac{1}{R_1} - \frac{1}{R_2} \right) $$
<br>


```python
n = 1.5   # indice di rifrazione

R1 = input("Inserire raggio curvatura R1:\n")   # richiedo input all'utente
R1 = float(R1)                                  # converto stringa input in float

R2 = input("Inserire raggio curvatura R2:\n")
R2 = float(R2)

F = (n-1) * (1/R1 + 1/R2)
f=1/F

print(str(f))      # stampo il risultato
```

# Programmino che valuta l'azione di una lente (if-else incastonati)


Il programma chiede in input all'utente la focale della lente (sottile) e la distanza di un oggetto da essa. A seconda dei valori, stabilisce se l'immagine dell'oggetto verrà ingrandita o ridotta, se comparirà a destra o a sinistra della lente e se sarà o meno invertita. Anche se non fa calcoli, nello scrivere le condizioni noi abbiamo usato le due formule 
<br>
<br>
$$ \frac{1}{z_1} + \frac{1}{z_2} = \frac{1}{f} \hspace{15mm} y_2 = - \frac{z_2}{z_1} y_1$$
<br>
con riferimento alla figura
<br>


```python
f = input("Inserire la focale f:\n")      # richiedo input all'utente
f = int(f)                                 

z1 = input("Inserire la distanza dell'oggetto z1:\n")   
z1 = int(z1)

if (f > 0):      # inizio asserzioni condizionate concatenate
    
    if (z1 > 2*f):
        print ("Lente: convessa.   Immagine: invertita e ridotta.")
        print ("Posizione immagine: distanza tra f e 2f sulla destra della lente")
        
    elif (z1 == 2*f):
        print ("Lente: convessa.   Immagine: invertita ma stessa dimensione.")
        print ("Posizione immagine: distanza 2f sulla destra della lente")
        
    elif (f < z1 < 2*f):
        print ("Lente: convessa.   Immagine: invertita ed ingrandita.")
        print ("Posizione immagine: distanza >2f sulla destra della lente")
        
    elif (z1 == f):
        print ("Lente: convessa.   Immagine: nessuna.")
        print ("Posizione immagine: all'infinito")
        
    elif (z1 < f):
        print ("Lente: convessa.   Immagine: dritta ed ingrandita.")
        print ("Posizione immagine: a sinistra della lente")
        
else:
        print ("Lente: concava.   Immagine: dritta e ridotta.")
        print ("Posizione immagine: a sinistra della lente")
```

# Programmino che calcola il quadrato di un intero (ciclo while)

Il programma calcola il quadrato di un numero x sommando x stesso per x volte. Serve ad illustrare un ciclo iterativo.

```python
a = input("Inserire un numero intero x:\n")   # richiedo input all'utente
a = int(a)  

ans = 0       # inizializzo la variabile ans

iterLeft = a     #stabilisco il numero di iterazioni (contatore che andrà diminuendo ad ogni ciclo)

while (iterLeft != 0):         # condizione per la fine del ciclo
    ans = ans + a
    iterLeft = iterLeft - 1 
    
print ("Il quadrato di "  + str(a)  + " è "  + str(ans))
```

# Programmino che calcola il fattoriale di un intero (ciclo for)

```python
n= input("Inserire un numero intero n:\n")   # richiedo input all'utente
n = int(n)  

ans = 1       # inizializzo la variabile ans

for i in range(1, n+1):         # range(n) = [0, 1, 2...., n-1]   mentre   range(a,b) = [a, a+1, a+2,...., b-1]
    ans = ans * i
    
print ("Il fattoriale di "  + str(n)  + " è "  + str(ans))
```

Il programma chiede in input un intero e ne cerca una fattorizzazione; se non la trova avverte che si tratta di un numero primo. Serve ad illustrare l'uso del comando **break**.

```python
n= input("Inserire un numero intero n:\n")   # richiedo input all'utente
n = int(n) 

for x in range(2, n):
         if n % x == 0:
            print(n, 'equals', x, '*', n//x)
            break                               
else:                                               #notare che 'else' chiude il ciclo for, non il ciclo if
    # il ciclo ha finito senza trovare un fattore
        print(n, 'è un numero primo')
```

# Programmino  con funzione
Creiamo una **funzione** che chiede un intero n e stampa i numeri di Fibonacci fino ad n.

```python
def fib(n):    # scrive una sequenza Fibonacci fino a n
    """stampa una lista contenente la serie Fibonacci fino a n"""
    a, b = 0, 1
    while a < n:
        print(a, end=' ')
        a, b = b, a+b
    print()

    
N = input("Inserire un numero intero N:\n")   # richiedo input all'utente
N = int(N) 

 # Ora chiamiamo la nostra funzione:
fib(N)



```

# Programmino conversione Libbre - Kg

```python
peso = input("Inserire il proprio peso: ")
peso = float(peso)

unita_misura_peso = input("Indicare se il peso inserito è in Libbre (L) o in chilogrammi (K): ")


if unita_misura_peso.upper() == 'L':                # Nota che per confrontare con una stringa ho dovuto metterla tra virgolette. Il .upper() serve a far sì che se l'utente inserisce una lettera minuscola il programma funzioni ugualmente. 
    peso_kg = peso * 0.45
    print(f'Il tuo peso in Kg è : {peso_kg}')    # per stampare ho usato una stringa formattata
    
elif unita_misura_peso.upper() == 'K':
    peso_libbre = peso * 2.2
    print(f'Il tuo peso in libbre è : {peso_libbre}')

else:
    print("Non ho capito l'unita di misura!")
```

# Guessing game

Il programma chiede all'utente di indovinare un numero intero segreto (compreso tra 1 e 10) e gli dà 3 possibilità per indovinare.

```python
num_segreto = 9

guess_counter = 0
guess_utente = int(input("Indovinare il numero segreto: "))

while guess_counter < 2 :
    
    guess_utente = int(input("Sbagliato! Riprova: "))
    guess_counter += 1

    if guess_utente == num_segreto:
        print(f'Hai indovinato! Il numero segreto è: {guess_utente}')
        break
        
else:                                  # notare che qui l'ELSE chiude il WHILE, non l'IF
    print("Mi spiace ma hai esaurito i tentativi")
          
```

## Randomizzazione a blocchi per Trottola

```python
import numpy as np
import itertools
import csv
import math
import random
import scipy
from numpy import pi,sqrt,exp
```

```python
# genero tutte le combo con itertools

N = 140   # numero di pazienti
n = 4       # lunghezza di ciascun blocco 
m = int(n/2) 
num_blocks = int(N/n)

# N = math.comb(n,m)   #coefficiente binomiale di n/2 ed n
print(N)

combos_indexes = range(1,n+1)
combos = list(itertools.combinations(combos_indexes,m))
# combos_printable=list(map(list,combos)) 
# print(combos)

non_flat_list = []
for i in range(1, num_blocks+1):
    non_flat_list.append(random.choice(combos))
    
# print(non_flat_list)

non_flat_list_letters = []

for combo in non_flat_list:
    temp = ['C'] * n
    
    for j in combo:  
        temp[j-1]='T'
    
    non_flat_list_letters.append(temp)
        
# print(non_flat_list_letters)

flat_list_letters = np.array([item for sublist in non_flat_list_letters for item in sublist])
list_numbers = np.arange(1,N+1)


# stampare la tabella 

# row_format ="{:>15}" * 2
# print(row_format.format("Paziente n°", "T/C"))
# for num, lett in zip(range(N), flat_list_letters):
#     print(row_format.format(num, *lett))

from tabulate import tabulate
col_headers = ["Paziente n°", "Trattamento/Controllo"]

merged_array = np.array([list_numbers, flat_list_letters]).T

table = tabulate(merged_array , col_headers, tablefmt="fancy_grid", floatfmt = (".0f",".13f"))

print(table)

import csv

with open('Randomizzazione_Trottola.csv', 'w', newline='') as csvfile:
    fieldnames = ['Paziente n°', 'Trattamento/Controllo']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames, delimiter='\t')
    writer.writeheader()
    for i, j in zip(list_numbers, flat_list_letters):
        writer.writerow({'Paziente n°': i, 'Trattamento/Controllo': j})
```
