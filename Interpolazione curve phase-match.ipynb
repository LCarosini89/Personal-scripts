```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate, scipy.optimize

x  = np.array([35, 40, 45, 50, 55, 60])
y1=np.array([0.8135, 0.8123, 0.8111, 0.8100, 0.8089, 0.8078])

def sigto_idler (l_onda):
    valore = 1/((1.0/0.405) - (1.0 / l_onda))                
    return valore

vet1=[]
for i in range(len(yi)):
        vet1.append(sigto_idler(y1[i]))
y2 = np.array(vet1)



plt.plot(x, y1, marker='o', mec='none', ms=4, lw=1, label='$\lambda_i$')
plt.plot(x, y2, marker='o', mec='none', ms=4, lw=1, label='$\lambda_s$')
plt.xlabel("Crystal temperature (°C)")


interp1 = scipy.interpolate.InterpolatedUnivariateSpline(x, y1)
interp2 = scipy.interpolate.InterpolatedUnivariateSpline(x, y2)


g=interp1(27)
h=interp2(27)

new_x = np.linspace(x.min(), x.max(), 100)
new_y1 = interp1(new_x)
new_y2 = interp2(new_x)



idx = np.argwhere(np.diff(np.sign(new_y1 - new_y2)) != 0)
plt.plot(new_x[idx], new_y1[idx], 'ro', ms=7, label='degenerate emission')


print(new_x[idx])
print(g)
print(h)

plt.legend(fontsize=15, bbox_to_anchor=(1.04,1), loc="upper left")
#plt.legend(frameon=False, fontsize=10, numpoints=1)

plt.savefig('curve crossing.png', dpi=200)
plt.show()
```
